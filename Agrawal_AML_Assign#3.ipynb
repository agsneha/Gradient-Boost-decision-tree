```python
from imblearn.over_sampling import SMOTE
```

```python
# Function to augment Wine dataset using SMOTE

def augment_wine_data(wine_df):
    # Split features and target
    X_wine = wine_df.drop(columns='target')
    y_wine = wine_df['target']
    
    datasets = {}
    datasets['Dataset 1'] = wine_df.copy()  # Original dataset

    # Creating augmented datasets
    for i in range(1, 5):
        # Calculate the number of samples to generate
        n_samples = len(wine_df) * i
        
        # Generating samples with SMOTE
        smote = SMOTE(random_state=42)
        X_resampled, y_resampled = smote.fit_resample(X_wine, y_wine)

        # Creating DataFrame from the resampled data
        augmented_df = pd.DataFrame(X_resampled, columns=X_wine.columns)
        augmented_df['target'] = y_resampled
        
        # Combining original dataset with the required amount of augmented data
        # Here, I am taking the original dataset and concatenating it with the first 'n_samples' of the augmented data
        datasets[f'Dataset {i + 1}'] = pd.concat([wine_df, augmented_df], ignore_index=True)

        # Updating the augmented_df to include more samples for the next iteration
        wine_df = datasets[f'Dataset {i + 1}'].copy()  # Updating to the newly created dataset for the next loop

    return datasets
```

```python
augmented_wine_datasets = augment_wine_data(wine_df)
```

```python
# Shapes of the augmented datasets
for key, value in augmented_wine_datasets.items():
    print(f"{key} (Wine Dataset): {value.shape}")
```

```python
#### DIABETES DATASET
```

```python
import numpy as np
```

```python
# Function to augment Diabetes dataset using Gaussian noise
def augment_diabetes_data(diabetes_df):
    datasets = {}
    datasets['Dataset 1'] = diabetes_df.copy()  # Original dataset

    # Creating 4 augmented datasets
    for i in range(1, 5):
        # Generating Gaussian noise with the same shape as the original dataset
        noise = np.random.normal(0, 0.2, diabetes_df.shape)
        augmented_df = diabetes_df + noise
        augmented_df = np.clip(augmented_df, 0, None)  # Assuming no negative values are expected

        # Combining with original dataset
        datasets[f'Dataset {i + 1}'] = pd.concat([datasets['Dataset 1'], augmented_df], ignore_index=True)
        
        # Updating diabetes_df for the next iteration, 
        # so that the augmented_df in the next iteration is built upon the original dataset
        diabetes_df = datasets[f'Dataset {i + 1}']

    return datasets
```

```python
augmented_diabetes_datasets = augment_diabetes_data(diabetes_df)
```

```python
for key, value in augmented_diabetes_datasets.items():
    print(f"{key} (Diabetes Dataset): {value.shape}")
```

```python
"""
Task 4: Implement XGBoost, LightBoost, and CatBoost.
"""
```

```python
import time
from xgboost import XGBClassifier, XGBRegressor
from catboost import CatBoostClassifier, CatBoostRegressor
from lightgbm import LGBMClassifier, LGBMRegressor
from sklearn.metrics import accuracy_score, r2_score
```

```python
"""Function to Evaluate and Time Models"""

def evaluate_model(model, X_train, y_train, X_test, y_test, is_classification=True):
    start_time = time.time()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    end_time = time.time()
    
    if is_classification:
        accuracy = accuracy_score(y_test, y_pred)
        return accuracy, end_time - start_time
    else:
        r2 = r2_score(y_test, y_pred)  # Make sure r2_score is correctly used as a function
        return r2, end_time - start_time
```

```python
"""Initializing Models for Classification and Regression"""
classifiers = {
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    'CATBoost': CatBoostClassifier(verbose=0, random_state=42),
    'LightGBM': LGBMClassifier(random_state=42)
}

regressors = {
    'XGBoost': XGBRegressor(random_state=42),
    'CATBoost': CatBoostRegressor(verbose=0, random_state=42),
    'LightGBM': LGBMRegressor(random_state=42)
}
```

```python
"""Initializing dictionaries to store results"""

wine_results = {}
diabetes_results = {}

# Dataset names
datasets = ['Dataset 1', 'Dataset 2', 'Dataset 3', 'Dataset 4', 'Dataset 5']
```

```python
"""Evaluating Classifiers on Wine Datasets"""

for dataset_name, dataset in augmented_wine_datasets.items():
    print(f"\nEvaluating models for {dataset_name} (Wine Dataset)")
    X_wine = dataset.drop(columns='target')
    y_wine = dataset['target']
    X_train, X_test, y_train, y_test = train_test_split(X_wine, y_wine, test_size=0.3, random_state=42)
    
    wine_results[dataset_name] = {}
    for model_name, model in classifiers.items():
        accuracy, exec_time = evaluate_model(model, X_train, y_train, X_test, y_test, is_classification=True)
        wine_results[dataset_name][model_name] = (accuracy, exec_time)
        print(f"{model_name} - Accuracy: {accuracy:.4f}, Time: {exec_time:.4f} seconds")
```

```python
"""Evaluating Regressors on Diabetes Datasets"""

for dataset_name, dataset in augmented_diabetes_datasets.items():
    print(f"\nEvaluating models for {dataset_name} (Diabetes Dataset)")
    X_diabetes = dataset.drop(columns='target')
    y_diabetes = dataset['target']
    X_train, X_test, y_train, y_test = train_test_split(X_diabetes, y_diabetes, test_size=0.3, random_state=42)
    
    diabetes_results[dataset_name] = {}
    for model_name, model in regressors.items():
        r2, exec_time = evaluate_model(model, X_train, y_train, X_test, y_test, is_classification=False)
        diabetes_results[dataset_name][model_name] = (r2, exec_time)
        print(f"{model_name} - R² Score: {r2:.4f}, Time: {exec_time:.4f} seconds")
```

```python
"""Function to Display Results in a Table Format"""

def display_results(results, dataset_type):
    print(f"\n{'='*30} {dataset_type} RESULTS {'='*30}")
    for dataset_name, models in results.items():
        print(f"\n{dataset_name}:")
        for model_name, (score, exec_time) in models.items():
            print(f"{model_name} - Score: {score:.4f}, Time: {exec_time:.4f} seconds")
    print(f"{'='*80}")
```

```python
# Results for Wine Dataset

display_results(wine_results, 'Wine Dataset')
```

```python
# Results for Diabetes Dataset

display_results(diabetes_results, 'Diabetes Dataset')
```
